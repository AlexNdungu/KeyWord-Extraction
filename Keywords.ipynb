```python
import pandas as pd
import re

#import html.parser as new_html
#Transform to html_parse
#html_parser = new_html.HTMLParser()

import html

from spacy.lang.en.stop_words import STOP_WORDS
```

```python
#Here we read the file
stackoverlow_data = pd.read_csv('train.csv')
```

```python
#Now we clean the dataset
def data_cleaning(text):
    #Removing Html tag
    def escaping_html_char(text):
        
        parse_text = html.unescape(text)
        return parse_text
    #Here we clean the data
    def clean(text):
        
        text = re.sub('[^\w\s]',"", text.lower())
        text = " ".join([word for word in text.split() if word not in STOP_WORDS])
        return text
    
    text = escaping_html_char(text)
    text = clean(text)
    return text
```

```python
#Here we create on column containing title and body named text
stackoverlow_data['text'] = stackoverlow_data['Title'] + ' ' + stackoverlow_data['Body']
```

```python
#Now we apply the datacleaning
stackoverlow_data['text_cleaned'] = stackoverlow_data['text'].apply(data_cleaning)
```
